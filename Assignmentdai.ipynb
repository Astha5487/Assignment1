```python
numeric_columns = ['Mean Scale Score', 'Num Level 1', 'Pct Level 1', 'Num Level 2', 
                   'Pct Level 2', 'Num Level 3', 'Pct Level 3', 'Num Level 4', 
                   'Pct Level 4', 'Num Level 3 and 4', 'Pct Level 3 and 4']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check for missing values in important columns
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Drop rows with missing 'Mean Scale Score' (if necessary)
df_clean = df.dropna(subset=['Mean Scale Score'])
```

* The missing values insight shows that:

Key Gaps: Mean Scale Score has the most missing values (91,882), while other columns like proficiency levels (Num Level 1-4, Pct Level 1-4, and combined levels) are missing in around 48,000 entries.

Implications: These gaps may affect the analysis of student performance, particularly in assessing proficiency and scale scores, and require handling (e.g., imputation or exclusion) to maintain data quality for analysis.

Complete Data: Certain columns like Report Category, Geographic Subdivision, Grade, and Number Tested have no missing values, ensuring full coverage for these aspects.








```python
def plot_categorical_distributions(column):
    plt.figure(figsize=(6, 4))
    sns.countplot(y=column, data=df_clean, order=df_clean[column].value_counts().index)
    plt.title(f"Frequency Distribution of {column}")
    plt.show()

# Frequency distribution for 'Grade'
plot_categorical_distributions('Grade')

# Frequency distribution for 'Report Category'
plot_categorical_distributions('Report Category')

# Frequency distribution for 'Student Category'
plot_categorical_distributions('Student Category')

# Frequency distribution for 'Year'
plot_categorical_distributions('Year')
```

```python
# Visualize missing values using a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
```

* The provided heatmap illustrates the distribution of missing values across various features in a dataset. Darker colors indicate a higher proportion of missing data. Key observations include a substantial number of missing values in columns such as "Grade", "Year", and "Student Category".
```python
# Compute the correlation matrix for numerical columns
numeric_cols = ['Number Tested', 'Mean Scale Score', 'Pct Level 1', 'Pct Level 2', 
                'Pct Level 3', 'Pct Level 4', 'Pct Level 3 and 4']

correlation_matrix = df[numeric_cols].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

* The correlation matrix shows that higher mean scale scores are strongly associated with higher percentages of students in Levels 3 and 4 (positive correlations with +0.79 and +0.82), while negatively correlated with lower-performing students in Levels 1 and 2 (-0.76 and -0.57). Pct Level 1 and Pct Level 2 also have strong negative correlations with Pct Level 3 and 4, indicating that schools with more students in lower levels tend to have fewer high performers. This highlights a clear performance divide between higher and lower levels
